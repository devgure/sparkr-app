```python
# AI Training Notebook
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib

# Simulated dataset
data = pd.DataFrame({
    'age_diff': [2, 5, 1, 8, 3],
    'same_gender_pref': [1, 0, 1, 0, 1],
    'distance_km': [5, 50, 12, 100, 30],
    'common_interests': [4, 1, 3, 0, 2],
    'matched': [1, 0, 1, 0, 1]
})

X = data[['age_diff', 'same_gender_pref', 'distance_km', 'common_interests']]
y = data['matched']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

joblib.dump(model, 'model/compatibility_model.pkl')
print("Model trained and saved!")
```
